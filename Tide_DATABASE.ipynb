```python
#Task 4.1
import sqlite3
connection = sqlite3.connect("data/tide.db")

with open("data/TIDES.txt", "r") as file:
    data = [line.strip("\n").split("\t") for line in file]
    query = "INSERT INTO Tide(Date,Time, isHigh, Height) VALUES(?,?,?,?)"
    for i in range(len(data)):
        if data[i][2] == "HIGH":
            data[i][2] = 1
        else:
            data[i][2] = 0
        connection.execute(query, (data[i]))
        connection.commit()
    connection.close()
```
